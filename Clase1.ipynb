```python
import math

def calcular_raiz_cuadrada(numero):
  """
  Calcula la raíz cuadrada de un número.

  Args:
    numero: El número al que se le calculará la raíz cuadrada.

  Returns:
    La raíz cuadrada del número.
  """
  if numero < 0:
    return "No se puede calcular la raíz cuadrada de un número negativo."
  else:
    return math.sqrt(numero)

# Ejemplo de uso:
numero = 16
raiz = calcular_raiz_cuadrada(numero)
print(f"La raíz cuadrada de {numero} es: {raiz}")

numero_negativo = -4
raiz_negativa = calcular_raiz_cuadrada(numero_negativo)
print(raiz_negativa)
```
